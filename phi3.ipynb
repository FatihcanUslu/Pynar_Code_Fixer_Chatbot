````python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline

torch.random.manual_seed(0)

model = AutoModelForCausalLM.from_pretrained(
    "microsoft/Phi-3-mini-128k-instruct",
    device_map="cuda",
    torch_dtype="auto",
    trust_remote_code=True,
)
````

````python
tokenizer = AutoTokenizer.from_pretrained("microsoft/Phi-3-mini-128k-instruct")
````

````python
!pip install googletrans==4.0.0rc1
#https://pypi.org/project/googletrans/
from googletrans import Translator
translator = Translator()
````

````python
def metni_bol(metin):
    global bolunmus_metin
    bolunmus_metin = metin.split('```')
    return bolunmus_metin
````

````python
@anvil.server.callable
def modele_input_yolla1(hata_girdisi):
  messages=[
      { 'role': 'user', 'content': f"""
      This code has errors. Can you explain the error end show the fixed version of the code:
        {hata_girdisi}
      """}
  ]

  pipe = pipeline(
      "text-generation",
      model=model,
      tokenizer=tokenizer,
  )

  generation_args = {
      "max_new_tokens": 500,
      "return_full_text": False,
      "temperature": 0.0,
      "do_sample": False,
  }
  #Modelden dönen sonucun eldesi
  output = pipe(messages, **generation_args)
  # print(output[0]['generated_text'])

  #Model sonucunu metin ve kod parcalarina bolme
  english_sonuc = metni_bol(output[0]['generated_text'])

  #Metin kısımlarını türkçeye çevirme
  turkce_sonuc = []

  for i, eleman in enumerate(english_sonuc):
          if i % 2 == 0:
              ceviri_sonucu = translator.translate(eleman, src="en", dest="tr")
              turkce_sonuc.append(ceviri_sonucu.text)
          else:
              turkce_sonuc.append(eleman)

  turkce_sonuc_str = "\n".join(turkce_sonuc)
  print("turkce olarak üretilen output:", """
  {}
  """.format(turkce_sonuc_str))
  # print(turkce_sonuc_str)
  return turkce_sonuc # Elde edilen sonuç geri döndürülmeyecek. Onun yerine liste döndürülecek, bu sayede elde edilmiş kod kısmı arayüz içinde kullanılabilir.
````
